```python
# Import the necessary libraries
import torch
import numpy as np
```

Tensors can be intitialized in several ways. One of which being directly from data, where the data type is automatically inferred.

```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
```

Another way is from a numpy array

```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
```

They can also be made from other tensors

```python
x_ones = torch.ones_like(x_data) # retains the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype = torch.float) # overrides the datatype of x_data
print(f"Random Tensor: \n {x_rand} \n")
```

Shape can be used to determine the dimesionality of an output tensor. Shape, itself is a tuple comprised of tensor dimensions.

```python
shape = (2, 3)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)


print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor} \n")
```

```python
# Conducting some operations on our tensor
tensor = torch.ones(4, 4)
print(f"First row: {tensor[0]}") # Just prints the first row of the tensor
print(f"First column: {tensor[:, 0]}") # Prints the first column of the tensor
print(f"Last column: {tensor[..., -1]}") # Prints the last column of the tensor; ... captures all row entries for the column
tensor[:, 1] = 0 # Sets the value of the second column of the tensor to 0
print(tensor)
```

Allegedly, tensors on the cpu and numpy arrays can share underlying memory locations; changing one will change the other.

```python
t = torch.ones(5)
print(f"t: {t}")
n= t.numpy() # Converts the tensor to a NumPy array
print(f"n: {n}")
```

Additionally, a change in the tensor will result in a change in the numpy array

```python
t.add_(1)
print(f"t: {t}")
print(f"n: {n}")
```

We can also show that changes in the numpy array result in changes in the tensor

```python
n = np.ones(5)
t = torch.from_numpy(n)

np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
